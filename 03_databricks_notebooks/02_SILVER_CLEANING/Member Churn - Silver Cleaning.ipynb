```python
-- 62_SILVER_CLEANING.SQL
-- Layer: Silver (Cleaned & Transformed)
-- Purpose: Clean, remove duplicates, and rename tables from Bronze.
-- Drop sqlite prefix. Filter null columns.

-- Step 1: Create Silver Schema
CREATE SCHEMA IF NOT EXISTS workspace.silver;

-- Step 2: Data Profiling
-- Check for NULL values
SELECT
    COUNT(*) AS total_rows,
    COUNT(*) - COUNT(member_id) AS null_member_ids,
    COUNT(*) - COUNT(persona) AS null_personas,
    COUNT(*) - COUNT(churned) AS null_churned
FROM workspace.bronze.sqlite_members;

-- Step 3: Clean and Load Silver Table
CREATE OR REPLACE TABLE workspace.silver.members AS
SELECT *
FROM workspace.bronze.sqlite_members
WHERE member_id IS NOT NULL
AND persona IS NOT NULL
AND churned IS NOT NULL;
```
